## A DataFrame has
- rows
- columns
- indexes
- data types

*Like Excel tables or SQL tables.*

````python
import pandas as pd

data = {
    "name":["Kabilesh","Dhanush","Vijay"],
    "age":[23,21,30],
    "grade":["A","B","A"]
}

df = pd.DataFrame(data)
print(df)
````

````python
print(df.head()) # View first Methods
````

````python
print(df.tail()) # view last rows
````

````python
print(df.shape)
````

````python
print(df.columns)
````

````python
print(df.dtypes)
````

````python
print(df.describe())
````

````python
print(df["name"])
````
