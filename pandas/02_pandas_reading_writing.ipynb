# Reading and Writing Data

`File -> Pandas DataFrame -> Clean -> Analyze -> Save`

Most data comes from:
- CSV files
- Excel files
- databases
- APIs

*CSV* = Comma-Separated Values
```text
name,age,gpa
Alice,21,3.8
Bob,22,3.4
Charlie,20,3.9
```
Each:

- line = row
- comma = separator
- first line = column names

````python
import pandas as pd

df = pd.read_csv("pandas/datasets/data.csv")
print(df)
````

`pd.read_csv("file.csv")`

This:
1. opens file
2. reads CSV content
3. converts it into a DataFrame

### Useful for:

| Method       | Purpose          |
| ------------ | ---------------- |
| `read_csv()` | Read CSV         |
| `to_csv()`   | Save CSV         |
| `head()`     | First rows       |
| `tail()`     | Last rows        |
| `info()`     | Dataset summary  |
| `shape`      | Rows and columns |
| `dtypes`     | Data types       |
