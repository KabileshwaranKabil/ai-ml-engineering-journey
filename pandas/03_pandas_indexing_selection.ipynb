# Indexing, Selection and Filtering
Most real Pandas work is:

- selecting rows
- filtering data
- querying conditions

## Two Core Indexing Tools
Pandas provides two main indexers:
| Method   | Meaning                       |
| -------- | ----------------------------- |
| `loc[]`  | label-based access            |
| `iloc[]` | integer-position-based access |


```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [21, 22, 23, 24],
    "gpa": [3.8, 3.4, 3.9, 3.2]
})

print(df)
```

```python
df.iloc[0] # select first row
```

```python
df.iloc[0,0] # select specific cell
```

```python
df.iloc[0:2] # select multiple rows
```

```python
df.iloc[0:2,0:2]
```

### loc — Label-Based Selection

```python
df.loc[:,"name"]
```

```python
df.loc[:,["name","gpa"]]
```

```python
# Select rows by condition
df.loc[df["age"]>22]
```

### Boolean Filtering

```python
df[df["age"]>21]
```

### Multiple Condition

```python
# AND Condition
df[(df["age"] >21) & (df["gpa"]>3.5)]
```

Rules:
- use & not and
- wrap each condition in parentheses

**Python and does not work with Pandas Series**

```python
# OR Condition
df[(df["age"] > 23) | (df["gpa"] > 3.8)]
```

### Filtering Text Data

```python
df[df["name"]=="Alice"]
```

```python
df[df["name"] !="Bob"]
```

```python
df[df["name"].isin(["Alice","Charlie"])]
```

### Filtering with NOT condition

```python
df[~(df["age"] > 22)]
```

### Resetting Index after Filtering

```python
# After filtering
filtered = df[df["age"] > 21]
print(filtered)
```

```python
filtered = filtered.reset_index(drop=True)
print(filtered)
```

| Pandas     | SQL                     |
| ---------- | ----------------------- |
| df[...]    | WHERE                   |
| loc        | SELECT columns + filter |
| conditions | WHERE clause            |

### Summary Table
| Operation           | Syntax                  |   |
| ------------------- | ----------------------- | - |
| First row           | `iloc[0]`               |   |
| Cell                | `iloc[row, col]`        |   |
| Column              | `loc[:, "col"]`         |   |
| Filter rows         | `df[df["col"] > value]` |   |
| Multiple conditions | `&` and `               | ` |
| Multiple values     | `isin()`                |   |
