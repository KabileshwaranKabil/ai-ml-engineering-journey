```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
```

```python
print(y)
```

```python
print(z)
```

### Plotting the data

```python
# Sine Wave
plt.plot(x,y)
plt.show()
```

```python
# Cosine Wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x axis and y axis labels
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("Sine Wave")
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+') # fill the line with red color +
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```
